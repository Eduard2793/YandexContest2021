```python
def F(x0,C):
    x1 = []
    x2 = []
    for i in range(len(x0)):
        x1.append(x0[i]**2)
    for i in range(len(x0)-1):
        x2_i = []
        for j in range(1,len(x0)-i):
            x2_i.insert(0, (x0[i]*x0[-j]))
            
        x2 += x2_i
            
    x = x0 + x1 + x2
    
    y = 0
    for i in range(len(x)):
        y += x[i]*C[i]
        
    return y
```
