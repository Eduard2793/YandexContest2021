```python
import numpy as np
import numpy.random as r
```

```python
def generate1():  
    a = r.uniform(0, 1)  
    b = r.uniform(0, 1)  
    return (a * np.cos(2 * np.pi * b), a * np.sin(2 * np.pi * b))
```

```python
def generate2():  
    while True:  
        x = r.uniform(-1, 1)  
        y = r.uniform(-1, 1)  
        if x ** 2 + y ** 2 > 1:  
            continue  
        return (x, y)
```

```python
X1 = []
Y1 = []
X2 = []
Y2 = []
a1 = []
a2 = []
for i in range(1000):
    (x1,y1) = generate1()
    (x2,y2) = generate2()
    a1.append((x1,y1))
    a2.append((x2,y2))
    X1.append(x1)
    Y1.append(y1)
    X2.append(x2)
    Y2.append(y2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X1,Y1,c='r')
plt.show()
plt.scatter(X2,Y2,c='g')
plt.show()
```

```python
def countCloseToZero(a):
    n = len(a) #1000
    k = 0
    for (x,y) in a:
        if (x**2 + y**2)**0.5 <= 0.25:
            k += 1
    return k/n
```

```python
kn1 = countCloseToZero(a1)
kn2 = countCloseToZero(a2)
```
